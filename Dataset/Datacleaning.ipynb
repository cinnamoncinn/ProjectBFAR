```python
# Ensure values are converted to integers before replacing
def to_int(value):
    try:
        return int(float(value))  # Converts '1.0' -> 1
    except:
        return value  # Return original if conversion fails

# Apply replacements
for col, mapping in replacements.items():
    if col in df.columns:
        df[col] = df[col].map(lambda x: mapping.get(to_int(x), x))

# Debug: Print unique values after replacement
for col in replacements.keys():
    if col in df.columns:
        print(f"Unique values after replacing {col}: {df[col].unique()}")
```
